### Visualization
This notebook is used to create visualizations of the first dataset created without policies.
The only relevant outcomes in this case are the deaths per location, and damage, as there are no costs yet.

```python
#This cell load in libraries and data
from ema_workbench import load_results
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

experiments, outcomes = load_results('full_results_exploration.tar.gz')
```

```python
deaths = pd.DataFrame(outcomes, columns = ['A.1_Expected Number of Deaths', 'A.2_Expected Number of Deaths', 'A.3_Expected Number of Deaths',
                                         'A.4_Expected Number of Deaths', 'A.5_Expected Number of Deaths'])
# plt.subplot(121)
# sns.violinplot(data=data, y='A.1_Expected Number of Deaths')
# plt.subplot(122)
# sns.violinplot(data=data, y='A.2_Expected Number of Deaths')
```

A violin plot is used as this shows the range and spread of outcomes in an easily interpretable way

```python
fig, ax = plt.subplots(figsize=(20,8))
sns.violinplot(data=deaths, scale='width');
```

```python
data = pd.DataFrame(outcomes)
sns.violinplot(data=data, y='Expected Annual Damage', color='gold')
low_damage = outcomes['Expected Annual Damage'] >= 1e8
damage = outcomes['Expected Annual Damage'] > 3e9
print('Fraction of solutions with more than 1e8 damage:',  sum(low_damage)/10000)
print('Fraction of solutions with more than 3e9 damage:',  sum(damage)/10000)
```

```python
deaths_damage = pd.DataFrame(outcomes, columns = ['A.1_Expected Number of Deaths', 'A.2_Expected Number of Deaths', 'A.3_Expected Number of Deaths',
                                         'A.4_Expected Number of Deaths', 'A.5_Expected Number of Deaths', 'Expected Annual Damage'])
sns.pairplot(pd.DataFrame(deaths_damage));
```
